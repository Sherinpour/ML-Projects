```python
# Histogram of numerical columns
df.hist(bins=20, figsize=(15, 10))
plt.suptitle("Distribution of Numerical Features")
plt.show()

# Boxplot to detect outliers
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    plt.figure(figsize=(6, 2))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
# Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Correlation heatmap

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

```

```python
# Boxplot of price vs airconditioning status
sns.boxplot(x='airconditioning', y='price', data=df)
plt.title('Price by Air Conditioning')
plt.show()
```

```python
# Boxplot of price vs furnishing status
sns.boxplot(x='furnishingstatus', y='price', data=df)
plt.title('🔹 Price by Furnishing Status')
plt.show()

# Boxplot of price vs mainroad status
sns.boxplot(x='mainroad', y='price', data=df)
plt.title('🔹 Price by Main Road Access')
plt.show()

# Boxplot of price vs guestroom status
sns.boxplot(x='guestroom', y='price', data=df)
plt.title('🔹 Price by Guest Room Availability')
plt.show()
```
