```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Global YouTube Statistics.csv", encoding="latin1")
```

```python
#cleaning data

for col in ['subscribers', 'video views', 'uploads', 'highest_monthly_earnings', 'lowest_monthly_earnings']:
    df[col] = df[col].astype(str).str.replace(',', '').str.replace(' ', '').str.extract('(\d+)', expand=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# (Univariate Analysis)
numeric_cols = ['subscribers', 'video views', 'uploads', 'highest_monthly_earnings']
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
# outliers with box plot
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python
#correlation between numeric features
plt.figure(figsize=(10, 6))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
categoric_cols = ['Country', 'category', 'channel_type']
for col in categoric_cols:
    print(f"\nTop values in '{col}':")
    print(df[col].value_counts().head(10))
```

```python
df.groupby(["category", "Country"])["subscribers"].sum().unstack().plot(kind="bar", stacked=True)
plt.title("subscibers breakdown by category and country")
plt.xlabel("category")
plt.ylabel("total subscribers")
plt.legend(title="country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.xticks(rotation  = 90)
plt.show()
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"], df["subscribers"], alpha=0.5)
plt.title("scatter plot of subscribers vs gross tertiary education enrollment")
plt.xlabel("Gross tertiary education enrollment")
plt.ylabel("subscribers")
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("avrage unemployment rate by channel type")
plt.xlabel("channel type")
plt.ylabel("avrage unemployment rate")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(["created_year","channel_type"])["subscribers"].mean().unstack().tail().plot(kind="bar")
plt.title("avrage subscribers by created year and channel type")
plt.xlabel("created year")
plt.ylabel("avrage subscribers")
plt.xticks(rotation=45)
plt.legend(title="channel type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
df.groupby(["created_year","created_month"])["subscribers"].sum().unstack().head().plot(kind="area")
plt.title("subscribers growth over time")
plt.xlabel("year")
plt.ylabel("total subscribers")
plt.xticks(rotation=45)
plt.legend(title="month", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
df.groupby(["created_year","Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind="line", marker="o")
plt.title("tend of the highest yearly earning by created year and country")
plt.xlabel("created year")
plt.ylabel("average highest yearly earning")
plt.xticks(rotation=45)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
